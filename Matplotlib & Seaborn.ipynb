```python
x1=[2,3,5,6]
y1=[5,8,10,3]
x2=[5,8,10,13]
y2=[2,3,5,6]
```

### Line Graph()

```python
plt.plot(x1,y1,color='g',marker='o',markersize=12,label='firstline',linestyle='--',linewidth=1)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x2,y2,color='r',marker='o',markersize=12,label='firstline',linestyle=':',linewidth=3)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x1,y1,color='b',marker='s',markersize=12,label='firstline',linestyle='--',linewidth=3)
plt.plot(x2,y2,color='r',marker='o',markersize=12,label='firstline',linestyle='-',linewidth=3)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()####
plt.legend()###show all label
plt.show()
```

```python
x1=[3,5,7,9]
y1=[12,15,18,8]
x2=[4,6,8,10]
y2=[15,17,19,10]
plt.bar(x1,y1,label='data1')
plt.bar(x2,y2,label='data2')
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

### Piechart
It is used in when the data is in percentage format it is an correct option to visulizse data.

```python
expvalues=[1400,600,300,410,250]
explabels=['rent','food','recharge','petrol','others']
plt.pie(expvalues,labels=explabels,startangle=180,wedgeprops={'edgecolor':'black'},autopct='%1.2f%%')###it is used for pie chart data
plt.axis('equal')
plt.show()
```

```python
m1=[1,2,3,4]
n1=[2,4,6,8]
m2=[2,4,6,8]
n2=[1,2,3,4]

plt.scatter(m1,n1,label='data1')
plt.scatter(m2,n2,label='data2')
plt.legend()
plt.show()
```

### Bivariate graph analysis--- two variables are used

```python
sns.lmplot(x='total_bill',y='tip',data=df1)##### lm= linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False)
plt.show()
```
